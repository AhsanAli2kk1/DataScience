```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\aa946\OneDrive\Desktop\DataScience\pandastutorial-main\pandastutorial-main\Datasets\university_enrollment_(Mini-project-01).csv')
df.head(20)
```

```python
df.info()
df.isnull()
df.isnull().sum()
df.describe()
```

```python
df['pre_score'] = df['pre_score'].replace('-', np.nan)
```

```python
df.isnull().sum()
```

```python
df['pre_requirement'] = df['pre_requirement'].fillna('Not Defined')
df.head(10)
```

```python
mean_value = df['post_score'].mean()
df['post_score'] = df['post_score'].fillna(mean_value)
df.head(20)
```

```python
df['pre_score'] = pd.to_numeric(df['pre_score'], errors='coerce')
mean_value = df['pre_score'].mean()
df['pre_score'] = df['pre_score'].fillna(mean_value)
df.head(24)
```

```python
df.isnull().sum()
df.describe()
```

```python
duplicates=df.duplicated(keep=False)
duplicate_rows = df[duplicates]
duplicate_rows
```

```python
sns.barplot(x='course_type', y='enrollment_count', data=df)
plt.title('Enrollment Count by Course Type')
plt.show()
```

```python
df['department'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Enrollment Distribution by Department')
plt.ylabel('')
plt.show()
```

```python
sns.lineplot(x='department', y='enrollment_count', data=df)
plt.title('Enrollment Count by Department')
plt.show()
```

```python
sns.barplot(x='department', y='enrollment_count', data=df)
plt.title('Enrollment Count by Department')
plt.xlabel('Department')
plt.ylabel('Enrollment Count')
plt.show()
```

```python
sns.barplot(x='year', y='enrollment_count', data=df)
plt.title('Enrollment Count by Year')
plt.show()
```

```python
sns.boxplot(x='year', y='department', data=df)
plt.title('department Count by Year')
plt.show()
```

```python
correlation_matrix = df[['pre_score', 'post_score', 'enrollment_count']].corr()
print(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df['course_type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Enrollment Distribution by Course Type')
plt.ylabel('')
plt.show()
```

```python
df['score_improvement'] = df['post_score'] - df['pre_score']
sns.barplot(x='course_type', y='score_improvement', data=df)
plt.title('Score Improvement by Course Type')
plt.show()
```

```python
df['score_improvement'] = df['post_score'] - df['pre_score']
sns.barplot(x='department', y='score_improvement', data=df)
plt.title('Score Improvement by Course Type')
plt.show()
```

```python
df['score_improvement'] = df['post_score'] - df['pre_score']
sns.boxplot(x='pre_requirement', y='score_improvement', data=df)
plt.title('Score Improvement by Course Type')
plt.show()
```
