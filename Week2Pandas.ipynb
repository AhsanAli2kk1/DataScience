```python
def transformationFunc(x):
    if x <= 50 and x >= 30:
        return x * 1.8
    elif x < 30:
        return x * 2.4
    else:
        return x

print(df.columns)
df['Physics'] = df['Physics'].apply(transformationFunc)
print("Updated DataFrame:")
print(df)
```

```python
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())

# Apply Min-Max scaling to the 'Physics' column
df['Physics_normalized'] = min_max_scaling(df['Physics'])
df
```

```python
mean=df['Physics'].mean()
std=df['Physics'].std()
df['Z-Normalization']=(df['Physics']-mean)/std
df
```

```python
def generateDate(StartDate,EndDate,NumDate):
    for i in range(NumDate):
        return StartDate+(EndDate-StartDate)*np.random.random()
StartDate=datetime.datetime(2024, 7, 1)  
EndDate=datetime.datetime(2024, 7, 31)
random_dates = generateDate(StartDate, EndDate, len(df))
df['Date'] = random_dates
df['Date'] = df['Date'].dt.date
print(df)
```
