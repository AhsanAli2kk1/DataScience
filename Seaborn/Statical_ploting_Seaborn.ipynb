```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

## Linear Regression Plot:

```python
sns.regplot(x='total_bill',y='tip',data=df)
plt.title('Linear Regression Plot of Total Bill vs Tip')
plt.show()
```

## Multiple Regression Plot:

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',data=df)
plt.show()
```

## Residual Plot:

```python
sns.residplot(x='total_bill',y='tip',data=df)
plt.show()
```

# Histrogram: 

```python
sns.histplot(data=df, x='total_bill', kde=True)
plt.title('Histogram of Total Bill with KDE')
plt.show()
```

## KDE Plot: 

```python
sns.kdeplot(x='tip',data=df, shade=True)
plt.show()
```

## Distribution Plot:

```python
sns.distplot(df['total_bill'])
plt.title('Distribution Plot of Total Bill')
plt.show()
```

## Box Plot:

```python
sns.boxplot(data=df, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

## Pair Plot:

```python
sns.pairplot(df)
plt.title('Pair Plot of Tips Dataset')
plt.show()
```

## Heatmap:

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Tips Dataset')
plt.show()
```

## Joint Plot:

```python
sns.jointplot(data=df,x='total_bill', y='tip', kind='hex')
plt.title('Joint Plot of Total Bill vs Tip')
plt.show()
```

## Facet Grid:

```python
g = sns.FacetGrid(df, col='smoker')
g.map(sns.histplot, 'total_bill')
plt.show()
```
