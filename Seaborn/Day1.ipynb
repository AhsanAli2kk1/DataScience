## Scatter Plot

```python
accidents_dep=df.groupby('department')['Work_accident'].sum().reset_index()
plt.figure(figsize=(12,4))
sns.scatterplot(x='department',y='Work_accident',data=accidents_dep)
```

```python
sns.scatterplot(x='time_spend_company',y='average_montly_hours',data=df)
```

```python
accidents_dep=df.groupby('department')['Work_accident'].sum().reset_index()
plt.figure(figsize=(12,4))
sns.lineplot(x='department',y='Work_accident',data=accidents_dep)
```

```python
accidents_dep=df.groupby('department')['Work_accident'].sum().reset_index()
plt.figure(figsize=(12,4))
sns.barplot(x='department',y='Work_accident',data=accidents_dep)
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='department',data=df)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='number_project',data=df)
```

```python
sns.boxplot(x='time_spend_company',y='average_montly_hours',data=df)
```

```python
tips = sns.load_dataset('tips')
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='total_bill')
plt.title('Line Plot of Total Bill by Size')
plt.show()
```

```python
sns.countplot(data=tips, x='day')
plt.title('Count Plot of Days')
plt.show()
```

```python
tips
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot with Whitegrid Style')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.set_palette('pastel')
sns.barplot(data=tips, x='day', y='total_bill')
plt.title('Bar Plot with Pastel Palette')
plt.show()
```
