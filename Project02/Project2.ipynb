```python
df['YOY_Sales'] = df['YOY_Sales'].str.replace('%', '').astype(float)
```

```python
df['YOY_Units'] = df['YOY_Units'].str.replace('%', '').astype(float)
```

```python
df.head(10)
```

```python
positive_count = df['YOY_Sales'][df['YOY_Sales'] > 0].count()
negative_count = df['YOY_Sales'][df['YOY_Sales'] < 0].count()
labels = ['Increase YOY Sales', 'Decrease YOY Sales']
explode = (0.1, 0) 
print(positive_count)
print(negative_count)
sizes = [positive_count, negative_count]
colors = ['#6666ff','#66b3ff']
plt.figure(figsize=(4, 4))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Percentage of Positive and Negative YOY Sales')
plt.show()
```

```python
df['Units'].sum()
```

```python
top_5_Units=df.groupby('Segment_Category')['Units'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_5_Units)
plt.subplot(2, 2, 1)
sns.barplot(x='Units',y='Segment_Category',data=top_5_Units)
plt.show()
```

```python
postive_units=df['YOY_Units'][df['YOY_Units']> 0].count()
negative_units=df['YOY_Units'][df['YOY_Units'] < 0].count()
labels = ['Increase YOY Sales', 'Decrease YOY Sales']
explode = (0.1, 0) 
sizes=[postive_units,negative_units]
print(postive_units)
print(negative_units)
colors=['#ff9999','#66b3ff']
plt.figure(figsize=(4, 4))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title('Percentage of Positive and Negative YOY Units')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()
print(correlation_matrix)
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Restuarant  Data')
plt.show()
```
